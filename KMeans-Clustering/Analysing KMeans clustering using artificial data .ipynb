```python
import pandas as dp
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Creating artificial data
from sklearn.datasets import make_blobs
```

```python
ar_data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
plt.scatter(ar_data[0][:,0],ar_data[0][:,1],c=ar_data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
n = 5
kmeans = KMeans(n_clusters=n)
```

```python
kmeans.fit(ar_data[0])
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('KMeans')
print(f'Number of Clusters:{n}')
ax1.scatter(ar_data[0][:,0],ar_data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(ar_data[0][:,0],ar_data[0][:,1],c=ar_data[1],cmap='rainbow')
```
