<h1>K-Means Clustering</h1>

```python
from sklearn.cluster import KMeans
```

```python
# Private - 1, Non-Private - 0
kmeans = KMeans(n_clusters=2)
```

```python
X = data.drop('Private',axis=1)
```

```python
X.head()
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

<h1> Evaluation </h1>

```python
data['Cluster'] = kmeans.labels_
```

```python
def converter(Cluster):
    if Cluster == 1:
        return 'Yes'
    else:
        return 'No'

data['Cluster'] = data['Cluster'].apply(converter)
```

```python
data[['Private','Cluster']]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
CM = confusion_matrix(data['Private'],data['Cluster'])
print(CM)
print('\n')
CR = classification_report(data['Private'],data['Cluster'])
print(CR)
```

```python

sns.lmplot(x="Grad.Rate", y="Room.Board", data=data,hue='Cluster',fit_reg=False)
sns.lmplot(x="Grad.Rate", y="Room.Board", data=data,hue='Private',fit_reg=False)
```
